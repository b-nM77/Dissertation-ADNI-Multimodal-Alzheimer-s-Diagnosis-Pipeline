
# ADNI Full Fusion — Diagnosis Pipeline (CogNID-style)

This notebook trains CN vs MCI vs Dementia diagnosis models on the **fused** feature table, mirroring the CogNID style while using your ADNI fusion.

**What's inside**
- Robust column detection (`PTID`, `Diagnosis`)
- Reproducibility: fixed seeds
- **Hybrid feature selection** (ANOVA + MI + RFE + RF voting) → `TOP_K_FEATURES`
- **Class imbalance** with **ADASYN** (applied **within** CV folds when available)
- Models: **RandomForest**, **SVM (RBF)**, **Stacking (RF+SVM→LR)**, and **XGBoost** (if installed)
- Metrics: **Accuracy**, **Macro-F1**, **Macro ROC-AUC (OvR)**, **Confusion Matrix**
- Plots: ROC curves, confusion matrices, feature importance & selection diagnostics
- CSV summary exported to `./diagnostics_out/full_fusion/`

> **Update the input path below** if your fusion file lives elsewhere.


```python

# === User config ===
from pathlib import Path
FUSION_XLSX = r"/Users/madhurabn/Desktop/adni/data/processed/fusion_master.xlsx"  # <--- set to your path
OUTDIR = Path("/Users/madhurabn/Desktop/adni/out/full_fusion")
OUTDIR.mkdir(parents=True, exist_ok=True)

# Optional force names (auto-detect runs if left as None)
FORCE_PTID_COL = None
FORCE_DIAG_COL = None

TOP_K_FEATURES = 60       # number of features after hybrid selection
N_FOLDS = 5
RANDOM_STATE = 42
SHOW_PLOTS = True

print("Using fusion file:", FUSION_XLSX)
print("Outputs directory:", OUTDIR.resolve())
```

```python

# === Imports & setup ===
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from pathlib import Path
import warnings, re

# plotting inline
%matplotlib inline

try:
    import seaborn as sns
    sns.set_context("notebook")
    _HAS_SNS = True
except Exception:
    _HAS_SNS = False

warnings.filterwarnings("ignore")

# sklearn, imbalanced-learn
from sklearn.model_selection import StratifiedKFold, cross_val_predict
from sklearn.preprocessing import OneHotEncoder, StandardScaler, label_binarize
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, confusion_matrix, classification_report
from sklearn.feature_selection import f_classif, mutual_info_classif, RFE
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.impute import SimpleImputer

# ADASYN (optional)
try:
    from imblearn.pipeline import Pipeline as ImbPipeline
    from imblearn.over_sampling import ADASYN
    _HAS_IMB = True
except Exception:
    _HAS_IMB = False

# XGBoost (optional)
try:
    from xgboost import XGBClassifier
    _HAS_XGB = True
except Exception:
    _HAS_XGB = False

np.random.seed(RANDOM_STATE)
```

```python

# === Utilities ===

def normalize_cols(cols):
    return [re.sub(r"\s+", " ", str(c)).strip() for c in cols]

def find_col(cols, candidates):
    cl = [c.lower() for c in cols]
    for tok in candidates:
        if isinstance(tok, str):
            for i, c in enumerate(cl):
                if tok == c or tok in c:
                    return cols[i]
        else:
            # callable predicate
            for i, c in enumerate(cols):
                if tok(c):
                    return c
    return None

def safe_name(s):
    return re.sub(r'[\\/*?:"<>| ]+', "_", str(s))

def detect_id_and_target(df, force_ptid=None, force_diag=None):
    cols = list(df.columns)
    if force_ptid and force_ptid in cols:
        ptid = force_ptid
    else:
        ptid = find_col(cols, ["ptid","rid","subjectid","subject id","id"])
    if force_diag and force_diag in cols:
        diag = force_diag
    else:
        diag = find_col(cols, ["diagnosis","dx","diag"])
    return ptid, diag

def map_diag_labels(s):
    # Map common encodings to CN/MCI/DEMENTIA (flexible)
    lbl = s.astype(str).str.strip().str.upper().replace({
        "1":"CN","2":"MCI","3":"DEMENTIA",
        "CN":"CN","CTL":"CN","CONTROL":"CN",
        "MCI":"MCI","EMCI":"MCI","LMCI":"MCI",
        "AD":"DEMENTIA","DEMENTIA":"DEMENTIA","ALZHEIMER":"DEMENTIA"
    })
    return lbl

def plot_confusion(y_true, y_pred, labels, title, out_png):
    cm = confusion_matrix(y_true, y_pred, labels=labels)
    plt.figure(figsize=(4.5,4))
    if _HAS_SNS:
        sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False,
                    xticklabels=labels, yticklabels=labels)
    else:
        plt.imshow(cm, cmap="Blues"); plt.colorbar()
        plt.xticks(range(len(labels)), labels); plt.yticks(range(len(labels)), labels)
    plt.title(title); plt.xlabel("Predicted"); plt.ylabel("True")
    plt.tight_layout(); plt.savefig(out_png, dpi=150)
    if SHOW_PLOTS: plt.show()
    plt.close()
```

```python

# === Load fusion & basic cleaning ===
fusion = pd.read_excel(FUSION_XLSX)
fusion.columns = normalize_cols(fusion.columns)

ptid_col, diag_col = detect_id_and_target(fusion, FORCE_PTID_COL, FORCE_DIAG_COL)
if ptid_col is None or diag_col is None:
    raise ValueError(f"Could not detect PTID/Diagnosis. PTID={ptid_col}, Diagnosis={diag_col}")

# Drop empty diagnosis rows
before = fusion.shape[0]
fusion = fusion[~fusion[diag_col].isna()].copy()
after = fusion.shape[0]
print(f"Dropped {before - after} rows with empty diagnosis. Remaining: {after}")

# Map diagnosis labels
fusion[diag_col] = map_diag_labels(fusion[diag_col])

# Keep a tidy copy
print("Detected PTID:", ptid_col, " | Diagnosis:", diag_col)
fusion.head(5)
```

```python
import re
import numpy as np
import pandas as pd

# --- 1) Identify/construct the target ----------------------------------------
# If you already have a single diagnosis column, set it here:
CANDIDATE_DIAG_COLS = ["Diagnosis", "diagnosis", "DX", "dx", "diag", "Diag"]
diag_col = next((c for c in CANDIDATE_DIAG_COLS if c in df.columns), None)

# If no single column, build it back from one-hot diagnosis_* columns:
if diag_col is None:
    onehot_diag_cols = [c for c in df.columns if re.fullmatch(r"(?i)diagnosis[_\- ]?(cn|mci|ad|dementia)", c)]
    if not onehot_diag_cols:
        raise ValueError("Couldn't find a diagnosis column or one-hot diagnosis_* columns.")
    # argmax across one-hots → class name
    class_names = []
    for c in onehot_diag_cols:
        if re.search(r'cn$', c, flags=re.I): class_names.append('CN')
        elif re.search(r'mci$', c, flags=re.I): class_names.append('MCI')
        elif re.search(r'(ad|dementia)$', c, flags=re.I): class_names.append('DEMENTIA')
        else: class_names.append(c)

    # Ensure consistent order of the one-hot columns
    order = ['CN','MCI','DEMENTIA']
    ordered_cols = []
    for name in order:
        for c, nm in zip(onehot_diag_cols, class_names):
            if nm == name:
                ordered_cols.append(c)
                break

    # argmax on ordered one-hots
    idx = df[ordered_cols].astype(float).values.argmax(axis=1)
    y_str = pd.Series([order[i] for i in idx], index=df.index, name="Diagnosis")
    df["Diagnosis"] = y_str
    diag_col = "Diagnosis"

# --- 2) Normalize string labels ----------------------------------------------
def normalize_diag(val):
    if pd.isna(val): return np.nan
    s = str(val).strip().upper()
    if s in {"CN","CONTROL","NORMAL"}: return "CN"
    if s in {"MCI"}: return "MCI"
    if s in {"AD","DEMENTIA","AD DEMENTIA","ALZHEIMER","ALZ"}: return "DEMENTIA"
    return s  # fallback (lets you see unexpected labels)

df[diag_col] = df[diag_col].map(normalize_diag)

# Drop rows with missing/unexpected diagnosis
valid = {"CN","MCI","DEMENTIA"}
bad_mask = ~df[diag_col].isin(valid)
if bad_mask.any():
    print(f"Dropping {bad_mask.sum()} rows with invalid diagnosis labels:", df.loc[bad_mask, diag_col].value_counts(dropna=False).to_dict())
df = df.loc[~bad_mask].copy()

# --- 3) Encode to integers in a fixed order -----------------------------------
CLASS_ORDER = ["CN","MCI","DEMENTIA"]
label_to_id = {lab:i for i, lab in enumerate(CLASS_ORDER)}
y = df[diag_col].map(label_to_id).astype(int).values  # -> 0,1,2

# Optional: keep a clean copy of y_str for reports/plots
y_str = df[diag_col].values

# --- 4) Build feature matrix X and ensure no target leakage -------------------
# Remove obvious ID/visit/diagnosis columns from features:
drop_cols = {diag_col, "PTID", "RID", "Subject", "visit", "VISCODE"}
# Also drop any one-hot diagnosis_* columns if they exist
drop_cols |= {c for c in df.columns if re.fullmatch(r"(?i)diagnosis[_\- ]?(cn|mci|ad|dementia)", c)}

X = df.drop(columns=[c for c in drop_cols if c in df.columns]).copy()

print("Target encoding:", label_to_id)
print("Shapes -> X:", X.shape, " y:", y.shape)
print("Class counts:", pd.Series(y_str).value_counts().to_dict())
```

```python

# === Train matrix & preprocessing ===
# Drop obvious non-features
drop_cols = [ptid_col]
maybe_drop = [c for c in ["visit","viscode","rid","subjectid"] if c in fusion.columns and c != ptid_col]
drop_cols += maybe_drop

X = fusion.drop(columns=[c for c in drop_cols if c in fusion.columns], errors="ignore")
y = fusion[diag_col].values

# Identify types
num_cols = [c for c in X.columns if np.issubdtype(X[c].dtype, np.number)]
cat_cols = [c for c in X.columns if c not in num_cols]

numeric_tf = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_tf = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

preproc = ColumnTransformer([
    ("num", numeric_tf, num_cols),
    ("cat", categorical_tf, cat_cols)
])

print(f"Feature columns -> numeric: {len(num_cols)}, categorical: {len(cat_cols)} | total: {X.shape[1]}")
```

```python

# === Hybrid feature selection: ANOVA + MI + RFE + RF voting ===
from sklearn.base import clone

# Step 1: pretransform to numeric matrix
X_proc = preproc.fit_transform(X)
# Build column names after OHE
ohe_cols = []
if cat_cols:
    ohe_cols = list(preproc.named_transformers_["cat"].named_steps["onehot"].get_feature_names_out(cat_cols))
all_feature_names = list(num_cols) + ohe_cols

# Encode y to integers for scorers
classes_ = np.unique(y)
class_to_int = {c:i for i,c in enumerate(classes_)}
y_int = np.array([class_to_int[v] for v in y])

# ANOVA
F, p = f_classif(X_proc, y_int)
rank_anova = pd.Series((-F).argsort().argsort(), index=all_feature_names)  # lower rank is better
rank_anova.name = "ANOVA_rank"

# Mutual information
MI = mutual_info_classif(X_proc, y_int, random_state=RANDOM_STATE)
rank_mi = pd.Series((-MI).argsort().argsort(), index=all_feature_names)
rank_mi.name = "MI_rank"

# RFE with linear model (one-vs-rest logistic for multi-class via liblinear)
est_rfe = LogisticRegression(max_iter=2000, multi_class="auto", solver="liblinear", class_weight="balanced", random_state=RANDOM_STATE)
rfe = RFE(estimator=est_rfe, n_features_to_select=max(10, min(TOP_K_FEATURES, X_proc.shape[1]//2)), step=0.1)
rfe.fit(X_proc, y_int)
rank_rfe = pd.Series((-rfe.ranking_).argsort(), index=all_feature_names)
rank_rfe.name = "RFE_rank"

# RandomForest importance
rf_tmp = RandomForestClassifier(
    n_estimators=400, max_depth=None, random_state=RANDOM_STATE, class_weight="balanced_subsample", n_jobs=-1
)
rf_tmp.fit(X_proc, y_int)
imp = getattr(rf_tmp, "feature_importances_", np.zeros(len(all_feature_names)))
rank_rf = pd.Series((-imp).argsort().argsort(), index=all_feature_names)
rank_rf.name = "RF_rank"

# Combine (average of z-scored ranks)
ranks = pd.concat([rank_anova, rank_mi, rank_rfe, rank_rf], axis=1)
ranks_z = (ranks - ranks.mean())/ranks.std(ddof=0)
score = -ranks_z.mean(axis=1)  # higher is better
rank_df = pd.DataFrame({
    "score": score,
    "ANOVA": rank_anova,
    "MI": rank_mi,
    "RFE": rank_rfe,
    "RF": rank_rf
}).sort_values("score", ascending=False)

selected_feats = rank_df.head(TOP_K_FEATURES).index.tolist()
rank_df.head(20)
```

```python

# === Plot top-k feature scores ===
plt.figure(figsize=(10, max(4, 0.28*len(selected_feats))))
rank_df.loc[selected_feats, "score"].sort_values().plot(kind="barh")
plt.title(f"Top-{TOP_K_FEATURES} hybrid-selected features")
plt.tight_layout(); plt.savefig(OUTDIR / "feature_selection_topk.png", dpi=150)
if SHOW_PLOTS: plt.show(); 
plt.close()

# Save selected list
pd.Series(selected_feats).to_csv(OUTDIR / "selected_features.csv", index=False)
print("Saved selected feature list ->", OUTDIR / "selected_features.csv")
```

```python

# === Build models ===
models = {}

models["RF"] = RandomForestClassifier(
    n_estimators=600, max_depth=None, min_samples_split=2, min_samples_leaf=1,
    random_state=RANDOM_STATE, class_weight="balanced_subsample", n_jobs=-1
)

models["SVM"] = SVC(
    kernel="rbf", C=2.0, gamma="scale", probability=True,
    class_weight="balanced", random_state=RANDOM_STATE
)

if _HAS_XGB:
    models["XGB"] = XGBClassifier(
        n_estimators=600, max_depth=4, learning_rate=0.05, subsample=0.9, colsample_bytree=0.9,
        reg_lambda=1.0, objective="multi:softprob", num_class=len(classes_), tree_method="hist",
        random_state=RANDOM_STATE
    )

# Stacking: base RF + SVM -> LR
base_estimators = []
base_estimators.append(("rf", models["RF"]))
base_estimators.append(("svm", models["SVM"]))
if _HAS_XGB:
    base_estimators.append(("xgb", models["XGB"]))

stack_final = LogisticRegression(max_iter=2000, class_weight="balanced", random_state=RANDOM_STATE)
models["Stack"] = StackingClassifier(
    estimators=base_estimators,
    final_estimator=stack_final,
    stack_method="predict_proba",
    passthrough=False, n_jobs=-1
)

list(models.keys())
```

```python

# === Cross-validated evaluation with ADASYN inside folds ===
from collections import defaultdict

skf = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=RANDOM_STATE)

results = []
proba_dict = {}
pred_dict = {}

for name, clf in models.items():
    if _HAS_IMB:
        pipe = ImbPipeline(steps=[
            ("preproc", preproc),
            ("adasyn", ADASYN(random_state=RANDOM_STATE, n_neighbors=5)),
            ("clf", clf)
        ])
    else:
        pipe = Pipeline(steps=[
            ("preproc", preproc),
            ("clf", clf)
        ])

    # Predict probabilities & labels via CV
    y_proba = cross_val_predict(pipe, X, y, cv=skf, method="predict_proba", n_jobs=-1)
    y_pred  = cross_val_predict(pipe, X, y, cv=skf, method="predict", n_jobs=-1)

    proba_dict[name] = y_proba
    pred_dict[name]  = y_pred

    acc  = accuracy_score(y, y_pred)
    f1m  = f1_score(y, y_pred, average="macro")
    # Macro ROC-AUC OvR
    y_bin = label_binarize(y, classes=classes_)
    try:
        aucm = roc_auc_score(y_bin, y_proba, multi_class="ovr", average="macro")
    except Exception:
        aucm = np.nan

    results.append({"Model": name, "Accuracy": acc, "MacroF1": f1m, "MacroROC_AUC_OvR": aucm})

res_df = pd.DataFrame(results).sort_values(by=["MacroROC_AUC_OvR","MacroF1","Accuracy"], ascending=False)
res_df
```

```python

# === Plots: confusion matrices & ROC ===
labels = list(classes_)

# Confusion matrices
for name in pred_dict:
    plot_confusion(y, pred_dict[name], labels, f"{name} — Confusion Matrix", OUTDIR / f"cm_{name}.png")

# ROC curves (OvR) — per class, compare models
y_bin = label_binarize(y, classes=labels)
plt.figure(figsize=(6,5))
for name, y_proba in proba_dict.items():
    # One-vs-rest macro ROC (average across classes)
    try:
        aucm = roc_auc_score(y_bin, y_proba, multi_class="ovr", average="macro")
        plt.plot([], [], label=f"{name} (macro AUC={aucm:.3f})")  # legend entry
    except Exception:
        plt.plot([], [], label=f"{name} (AUC N/A)")
plt.title("Macro ROC-AUC (OvR) — model comparison")
plt.legend()
plt.tight_layout(); plt.savefig(OUTDIR / "roc_macro_comparison.png", dpi=150)
if SHOW_PLOTS: plt.show()
plt.close()

# Save summary CSV
res_df.to_csv(OUTDIR / "model_cv_summary.csv", index=False)
print("Saved:", OUTDIR / "model_cv_summary.csv")
```

```python

# === Optional: RF feature importance (on full data with selected features) ===
# Refit a RF on the whole dataset pipeline to get importances post-preprocessing
rf_full = RandomForestClassifier(
    n_estimators=600, max_depth=None, random_state=RANDOM_STATE, class_weight="balanced_subsample", n_jobs=-1
)
pipe_full = Pipeline([("preproc", preproc), ("clf", rf_full)])
pipe_full.fit(X, y)

# Pull the transformed feature names
ohe_cols = []
if cat_cols:
    ohe = pipe_full.named_steps["preproc"].named_transformers_["cat"].named_steps["onehot"]
    ohe_cols = list(ohe.get_feature_names_out(cat_cols))
final_names = list(num_cols) + ohe_cols

importances = pipe_full.named_steps["clf"].feature_importances_
imp_ser = pd.Series(importances, index=final_names).sort_values(ascending=False)
topn = min(25, len(imp_ser))

plt.figure(figsize=(8, max(4, 0.3*topn)))
imp_ser.head(topn).sort_values().plot(kind="barh")
plt.title("RandomForest feature importance (top)")
plt.tight_layout(); plt.savefig(OUTDIR / "rf_feature_importance_top.png", dpi=150)
if SHOW_PLOTS: plt.show()
plt.close()

imp_ser.to_csv(OUTDIR / "rf_feature_importance_full.csv")
print("Saved RF importances ->", OUTDIR / "rf_feature_importance_full.csv")
```
