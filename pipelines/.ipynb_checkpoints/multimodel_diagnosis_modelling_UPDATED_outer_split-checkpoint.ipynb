
# ADNI Fusion Diagnosis Pipeline — ANOVA + ADASYN + RF / XGB / SVM / Stacking

This notebook loads the **fusion_master.xlsx**, applies **only ANOVA F-test** feature selection, handles class imbalance with **ADASYN**, and evaluates **Random Forest, XGBoost, SVM, and a Stacking Ensemble**.  
It mirrors CogNID-style structure while keeping your ADNI rules intact.

**Metrics:** Accuracy, macro-ROC AUC (OvR), Macro-F1, Confusion Matrix, per-class precision/recall/F1.  
**Plots:** ROC curves, confusion matrices, per-class metrics tables, top feature importances (RF & XGB), CV score bars.


```python

# ==== User configuration ====
from pathlib import Path
FUSION_XLSX = r"/Users/madhurabn/Desktop/adni/data/processed/fusion_master.xlsx"  # <--- set to your path
OUTDIR = Path("/Users/madhurabn/Desktop/adni/out/full_fusion")
OUTDIR.mkdir(parents=True, exist_ok=True)
PLOTDIR = OUTDIR / "plots"
PLOTDIR.mkdir(parents=True, exist_ok=True)

RANDOM_STATE = 42
N_FOLDS = 5
K_SELECTED = 60  # number of features to select via ANOVA (tune as needed)

print("Output dir:", OUTDIR.resolve())
print("Plots dir :", PLOTDIR.resolve())
```

```python

# ==== Imports ====
import warnings, numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
sns.set_context("notebook")
warnings.filterwarnings("ignore")

from sklearn.model_selection import StratifiedKFold, cross_val_score, cross_val_predict
from sklearn.metrics import accuracy_score, roc_auc_score, f1_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline as SkPipeline
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import ADASYN
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import SelectKBest, f_classif

from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

# XGBoost (optional)
try:
    from xgboost import XGBClassifier
    HAS_XGB = True
except Exception as e:
    HAS_XGB = False
    print("xgboost not available:", e)
```

```python

# ==== Load fusion master ====
df = pd.read_excel(FUSION_XLSX)
print("Loaded fusion shape:", df.shape)
print("Columns:", len(df.columns))
df.head(3)
```

```python

# ==== Identify PTID and Diagnosis; clean target ====
def find_col(cols, tokens):
    for c in cols:
        lc = str(c).strip().lower()
        if any(tok in lc for tok in tokens):
            return c
    return None

ptid_col = find_col(df.columns, ["ptid","subject","rid","id"])
diag_col = find_col(df.columns, ["diagnosis","dx","label","class"])

if ptid_col is None or diag_col is None:
    raise ValueError(f"Could not find PTID/Diagnosis. Detected PTID={ptid_col}, DIAG={diag_col}")

# drop empty diagnosis
before = df.shape[0]
df = df[df[diag_col].notna()].copy()
print(f"Dropped {before - df.shape[0]} rows with empty {diag_col}")

# Normalize diagnosis labels to {CN, MCI, DEMENTIA}
def norm_diag(x):
    s = str(x).strip().upper()
    if s in {"CN", "CONTROL", "COGNITIVELY NORMAL"}:
        return "CN"
    if s in {"MCI", "EMCI", "LMCI"}:
        return "MCI"
    if s in {"AD", "DEMENTIA", "ALZHEIMER", "ALZHEIMER'S DISEASE"}:
        return "DEMENTIA"
    return s

df[diag_col] = df[diag_col].apply(norm_diag)

# Encode to 0/1/2
label_map = {"CN":0, "MCI":1, "DEMENTIA":2}
y = df[diag_col].map(label_map)
if y.isna().any():
    bad = df.loc[y.isna(), diag_col].value_counts()
    print("Unmapped diagnosis values found:\n", bad)
    raise ValueError("Please fix unmapped diagnosis labels.")

print("Class counts:\n", y.value_counts().rename(index={0:"CN",1:"MCI",2:"DEMENTIA"}))
```

```python

# ==== Build feature matrix X (drop id/target, avoid leakage dummy columns) ====
drop_like = {ptid_col, diag_col}
# also drop any one-hot remnants of diagnosis if present
drop_like |= {c for c in df.columns if str(c).strip().lower().startswith(str(diag_col).strip().lower()+"_")}

meta_cols = list(drop_like)
X = df.drop(columns=meta_cols).copy()

# Identify numeric / categorical
num_cols = [c for c in X.columns if np.issubdtype(X[c].dtype, np.number)]
cat_cols = [c for c in X.columns if c not in num_cols]

print(f"Features total: {X.shape[1]} | numeric: {len(num_cols)} | categorical: {len(cat_cols)}")
```

```python

# ==== (NEW) Outer train/test split (80/20, stratified) ====
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, stratify=y, random_state=RANDOM_STATE
)

print("Train shape:", X_train.shape, "| Test shape:", X_test.shape)
```

```python

# ==== Preprocessor + ANOVA selector ====
pre = ColumnTransformer(
    transformers=[
        ("num", SkPipeline([("imp", SimpleImputer(strategy="median")), ("sc", StandardScaler())]), num_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore", sparse_output=False), cat_cols),
    ]
)

selector = SelectKBest(score_func=f_classif, k=min(K_SELECTED,  max(1, 5 if (len(num_cols)+len(cat_cols))<5 else len(num_cols)+len(cat_cols))))
```

```python

# ==== Helpers ====
def build_pipeline(clf):
    pipe = ImbPipeline(steps=[
        ("pre", pre),
        ("adasyn", ADASYN(random_state=RANDOM_STATE)),
        ("sel", selector),
        ("clf", clf),
    ])
    return pipe

cv = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=RANDOM_STATE)

labels = ["CN","MCI","DEMENTIA"]
inv_label_map = {v:k for k,v in {"CN":0,"MCI":1,"DEMENTIA":2}.items()}
```

```python

# ==== (UPDATED) Outer split + inner CV evaluation ====
import os, numpy as np, pandas as pd, re
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import cross_val_score, cross_val_predict, StratifiedKFold
from sklearn.metrics import roc_auc_score, confusion_matrix, classification_report, f1_score, accuracy_score

def evaluate_model(name, base_clf):
    """
    Now uses an outer 80/20 split (X_train/X_test), with 5-fold CV on the training set for model comparison,
    then trains on the full training set and reports final metrics on the held-out test set.
    Returns the *same keys as before* for CV metrics, and additional test_* keys for test metrics.
    """
    pipe = build_pipeline(base_clf)
    cv = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=RANDOM_STATE)

    # ---------- CV on TRAIN ----------
    acc_cv = cross_val_score(pipe, X_train, y_train, cv=cv, scoring="accuracy")
    f1m_cv = cross_val_score(pipe, X_train, y_train, cv=cv, scoring="f1_macro")

    y_proba_cv = cross_val_predict(pipe, X_train, y_train, cv=cv, method="predict_proba")
    y_pred_cv  = cross_val_predict(pipe, X_train, y_train, cv=cv, method="predict")
    auc_macro_ovr_cv = roc_auc_score(y_train, y_proba_cv, multi_class="ovr", average="macro")

    labels = ["CN","MCI","DEMENTIA"]
    cm_cv = confusion_matrix(y_train, y_pred_cv, labels=[0,1,2])
    rep_cv = classification_report(y_train, y_pred_cv, target_names=labels, output_dict=True)

    # ---------- Fit on full TRAIN and evaluate on TEST ----------
    pipe.fit(X_train, y_train)
    y_pred_test  = pipe.predict(X_test)
    y_proba_test = pipe.predict_proba(X_test)

    acc_test = accuracy_score(y_test, y_pred_test)
    f1m_test = f1_score(y_test, y_pred_test, average="macro")
    auc_macro_ovr_test = roc_auc_score(y_test, y_proba_test, multi_class="ovr", average="macro")
    cm_test = confusion_matrix(y_test, y_pred_test, labels=[0,1,2])
    rep_test = classification_report(y_test, y_pred_test, target_names=labels, output_dict=True)

    # ----- Print in dissertation-friendly wording -----
    print(f"\n=== {name} ===")
    print("[CV on training]")
    print(f"Accuracy (CV): {acc_cv.mean():.3f} ± {acc_cv.std():.3f}")
    print(f"Macro-ROC AUC (OvR, CV): {auc_macro_ovr_cv:.3f}")
    print(f"Macro-F1 (CV): {f1m_cv.mean():.3f} ± {f1m_cv.std():.3f}")
    print("Confusion Matrix (CV; rows=true, cols=pred):")
    print(cm_cv)

    print("\n[Held-out test]")
    print(f"Accuracy (test): {acc_test:.3f}")
    print(f"Macro-ROC AUC (OvR, test): {auc_macro_ovr_test:.3f}")
    print(f"Macro-F1 (test): {f1m_test:.3f}")
    print("Confusion Matrix (test; rows=true, cols=pred):")
    print(cm_test)

    # ----- Save artifacts -----
    safe_name = re.sub(r'[^A-Za-z0-9_.-]+', '_', name).strip('_')
    out_dir = PLOTDIR / safe_name
    out_dir.mkdir(parents=True, exist_ok=True)

    # CV confusion matrix
    plt.figure(figsize=(4.5,4))
    sns.heatmap(cm_cv, annot=True, fmt="d", cmap="Blues", xticklabels=labels, yticklabels=labels)
    plt.title(f"{name} — Confusion Matrix (CV on train)")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    cm_cv_path = out_dir / "confusion_matrix_cv.png"
    plt.tight_layout()
    plt.savefig(cm_cv_path, dpi=200); plt.show()
    print("Saved:", cm_cv_path)

    # TEST confusion matrix
    plt.figure(figsize=(4.5,4))
    sns.heatmap(cm_test, annot=True, fmt="d", cmap="Greens", xticklabels=labels, yticklabels=labels)
    plt.title(f"{name} — Confusion Matrix (Test)")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    cm_test_path = out_dir / "confusion_matrix_test.png"
    plt.tight_layout()
    plt.savefig(cm_test_path, dpi=200); plt.show()
    print("Saved:", cm_test_path)

    # Classification reports
    import pandas as pd
    pd.DataFrame(rep_cv).T.to_csv(out_dir / "classification_report_cv.csv")
    pd.DataFrame(rep_test).T.to_csv(out_dir / "classification_report_test.csv")
    print("Saved:", out_dir / "classification_report_cv.csv")
    print("Saved:", out_dir / "classification_report_test.csv")

    # Return CV keys as before + test keys
    return {
        "name": name,
        # CV metrics (same keys as before)
        "acc_mean": float(acc_cv.mean()),
        "acc_std": float(acc_cv.std()),
        "f1m_mean": float(f1m_cv.mean()),
        "f1m_std": float(f1m_cv.std()),
        "auc_macro_ovr": float(auc_macro_ovr_cv),
        # New test metrics
        "test_acc": float(acc_test),
        "test_macro_f1": float(f1m_test),
        "test_auc_macro_ovr": float(auc_macro_ovr_test)
    }
```

```python

# ==== Train & evaluate models (CV) ====
results = []

print("\n=== Random Forest ===")
rf = RandomForestClassifier(n_estimators=500, random_state=RANDOM_STATE, class_weight=None)
results.append(evaluate_model("Random Forest", rf))

if HAS_XGB:
    print("\n=== XGBoost ===")
    xgb = XGBClassifier(
        n_estimators=500,
        learning_rate=0.05,
        max_depth=4,
        subsample=0.8,
        colsample_bytree=0.8,
        objective="multi:softprob",
        eval_metric="mlogloss",
        random_state=RANDOM_STATE,
        num_class=3
    )
    results.append(evaluate_model("XGBoost", xgb))

print("\n=== SVM (RBF, prob=True) ===")
svm = SVC(kernel="rbf", probability=True, C=2.0, gamma="scale", random_state=RANDOM_STATE)
results.append(evaluate_model("SVM (RBF)", svm))

# Stacking (use simpler base models to keep runtime reasonable)
print("\n=== Stacking Ensemble ===")
base_estimators = [("rf", RandomForestClassifier(n_estimators=400, random_state=RANDOM_STATE)),
                   ("svm", SVC(kernel="rbf", probability=True, C=2.0, gamma="scale", random_state=RANDOM_STATE))]
if HAS_XGB:
    base_estimators.append(("xgb", XGBClassifier(
        n_estimators=300, learning_rate=0.05, max_depth=4,
        subsample=0.8, colsample_bytree=0.8, objective="multi:softprob",
        eval_metric="mlogloss", random_state=RANDOM_STATE, num_class=3)))

stack = StackingClassifier(
    estimators=base_estimators,
    final_estimator=LogisticRegression(max_iter=200, multi_class="ovr"),
    stack_method="predict_proba",
    passthrough=False, n_jobs=None
)

results.append(evaluate_model("Stacking", stack))

res_df = pd.DataFrame(results)
display(res_df)
res_df.to_csv(PLOTDIR / "model_cv_summary.csv", index=False)
```

```python

# ==== (NEW) Extract ANOVA-selected feature names and plot correlation heatmap ====
import numpy as np, pandas as pd
import matplotlib.pyplot as plt

# 1) Fit the preprocessor on full X (before CV) and transform
pre.fit(X, y)
X_pre = pre.transform(X)
# Ensure dense array for downstream steps
if hasattr(X_pre, "toarray"):
    X_pre = X_pre.toarray()

# 2) Get feature names after preprocessing (numeric + one-hot cat)
def get_feature_names_from_ct(ct):
    names = []
    for name, trans, cols in ct.transformers_:
        if name == "remainder":
            continue
        # If wrapped in a pipeline, try the last step
        if hasattr(trans, "named_steps"):
            last = list(trans.named_steps.values())[-1]
            if hasattr(last, "get_feature_names_out"):
                try:
                    out = last.get_feature_names_out(cols)
                except Exception:
                    out = last.get_feature_names_out()
                names.extend(list(out))
            else:
                names.extend(list(cols))
        else:
            if hasattr(trans, "get_feature_names_out"):
                try:
                    out = trans.get_feature_names_out(cols)
                except Exception:
                    out = trans.get_feature_names_out()
                names.extend(list(out))
            else:
                names.extend(list(cols))
    return list(map(str, names))

pre_feature_names = get_feature_names_from_ct(pre)

# 3) Fit the ANOVA selector on the preprocessed matrix
selector.fit(X_pre, y)
support_mask = selector.get_support()
selected_feature_names = [n for n, m in zip(pre_feature_names, support_mask) if m]

print(f"ANOVA selected features (n={len(selected_feature_names)}):")
for n in selected_feature_names:
    print(n)

# Save list to CSV
sel_path = OUTDIR / "anova_selected_features.csv"
pd.Series(selected_feature_names, name="feature").to_csv(sel_path, index=False)
print("Saved:", sel_path)

# 4) Correlation heatmap (Spearman) among selected features
sel_mat = X_pre[:, support_mask]
sel_df = pd.DataFrame(sel_mat, columns=selected_feature_names)
corr = sel_df.corr(method="spearman")

plt.figure(figsize=(max(6, len(selected_feature_names)*0.25), max(4, len(selected_feature_names)*0.25)))
plt.imshow(corr, interpolation="nearest")
plt.title("Correlation heatmap (Spearman) — ANOVA-selected features")
plt.colorbar()
plt.xticks(range(len(selected_feature_names)), selected_feature_names, rotation=90)
plt.yticks(range(len(selected_feature_names)), selected_feature_names)
plt.tight_layout()

heatmap_path = OUTDIR / "anova_selected_corr_heatmap.png"
plt.savefig(heatmap_path, dpi=200, bbox_inches="tight")
plt.show()
print("Saved:", heatmap_path)
```
