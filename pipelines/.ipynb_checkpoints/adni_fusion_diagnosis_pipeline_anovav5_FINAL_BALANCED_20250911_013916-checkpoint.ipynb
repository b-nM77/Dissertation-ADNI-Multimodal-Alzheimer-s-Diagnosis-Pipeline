
# ADNI Fusion Diagnosis Pipeline — ANOVA + ADASYN + RF / XGB / SVM / Stacking

This notebook loads the **fusion_master.xlsx**, applies **only ANOVA F-test** feature selection, handles class imbalance with **ADASYN**, and evaluates **Random Forest, XGBoost, SVM, and a Stacking Ensemble**.  
It mirrors CogNID-style structure while keeping your ADNI rules intact.

**Metrics:** Accuracy, macro-ROC AUC (OvR), Macro-F1, Confusion Matrix, per-class precision/recall/F1.  
**Plots:** ROC curves, confusion matrices, per-class metrics tables, top feature importances (RF & XGB), CV score bars.


```python

# ==== User configuration ====
from pathlib import Path
FUSION_XLSX = r"/Users/madhurabn/Desktop/adni/data/processed/fusion_master.xlsx"  # <--- set to your path
OUTDIR = Path("/Users/madhurabn/Desktop/adni/out/full_fusion")
OUTDIR.mkdir(parents=True, exist_ok=True)
PLOTDIR = OUTDIR / "plots"
PLOTDIR.mkdir(parents=True, exist_ok=True)

RANDOM_STATE = 42
N_FOLDS = 5
K_SELECTED = 60  # number of features to select via ANOVA (tune as needed)

print("Output dir:", OUTDIR.resolve())
print("Plots dir :", PLOTDIR.resolve())
```

```python

# ==== Imports ====
import warnings, numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
sns.set_context("notebook")
warnings.filterwarnings("ignore")

from sklearn.model_selection import StratifiedKFold, cross_val_score, cross_val_predict
from sklearn.metrics import accuracy_score, roc_auc_score, f1_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline as SkPipeline
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import ADASYN
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import SelectKBest, f_classif

from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

# XGBoost (optional)
try:
    from xgboost import XGBClassifier
    HAS_XGB = True
except Exception as e:
    HAS_XGB = False
    print("xgboost not available:", e)
```

```python

# === ADASYN (training-only, applied within training logic) ===
import numpy as np, pandas as pd, matplotlib.pyplot as plt
try:
    import seaborn as sns
    sns.set_context("talk"); sns.set_style("white")
except Exception:
    sns = None

from imblearn.over_sampling import ADASYN, SMOTE

_Xtr_name, _Ytr_name = None, None
for nm in ["X_train_scaled", "X_train"]:
    if nm in globals():
        _Xtr_name = nm; break
for nm in ["y_train"]:
    if nm in globals():
        _Ytr_name = nm; break
if _Xtr_name is None or _Ytr_name is None:
    raise NameError("Could not find X_train(_scaled) / y_train. Run splitting/scaling before this cell.")

_Xtr = globals()[_Xtr_name]
_ytr = np.asarray(globals()[_Ytr_name])

vals, cnts = np.unique(_ytr, return_counts=True)
labels = None
for nm in ["class_names", "label_names", "diagnosis_names", "classes"]:
    if nm in globals(): labels = globals()[nm]; break
labs = [labels[v] if (labels is not None and v < len(labels)) else str(v) for v in vals]
if sns is not None:
    plt.figure(figsize=(7,5)); sns.barplot(x=labs, y=cnts)
else:
    plt.figure(figsize=(7,5)); plt.bar(labs, cnts)
plt.title("Class Distribution — BEFORE resampling")
plt.xlabel("Class"); plt.ylabel("Count"); plt.xticks(rotation=15)
plt.tight_layout(); plt.show()

max_count = cnts.max()
sampling_strategy = {cls: int(max_count * 0.70) for cls, c in zip(vals, cnts) if c < (max_count / 1.5)}
min_count = cnts.min()
effective_neighbors = max(1, min(5, min_count - 1))

def _fit_resample(sampler, X, y):
    was_df = isinstance(X, pd.DataFrame)
    cols = X.columns if was_df else None
    Xr, yr = sampler.fit_resample(X, y)
    if was_df and not isinstance(Xr, pd.DataFrame):
        Xr = pd.DataFrame(Xr, columns=cols)
    return Xr, yr

_method = "none"
try:
    ada = ADASYN(sampling_strategy=sampling_strategy, n_neighbors=effective_neighbors, random_state=42)
    _Xr, _yr = _fit_resample(ada, _Xtr, _ytr)
    _method = "ADASYN"
except Exception as e:
    print(f"ADASYN failed ({e}). Falling back to SMOTE.")
    try:
        sm = SMOTE(sampling_strategy=sampling_strategy, k_neighbors=effective_neighbors, random_state=42)
        _Xr, _yr = _fit_resample(sm, _Xtr, _ytr)
        _method = "SMOTE"
    except Exception as e2:
        print(f"SMOTE also failed ({e2}). No resampling applied.")
        _Xr, _yr = _Xtr, _ytr

globals()[_Xtr_name] = _Xr
globals()[_Ytr_name] = _yr

vals2, cnts2 = np.unique(globals()[_Ytr_name], return_counts=True)
labs2 = [labels[v] if (labels is not None and v < len(labels)) else str(v) for v in vals2]
if sns is not None:
    plt.figure(figsize=(7,5)); sns.barplot(x=labs2, y=cnts2)
else:
    plt.figure(figsize=(7,5)); plt.bar(labs2, cnts2)
plt.title(f"Class Distribution — AFTER resampling ({_method})")
plt.xlabel("Class"); plt.ylabel("Count"); plt.xticks(rotation=15)
plt.tight_layout(); plt.show()

print(f"Training variables reassigned to balanced set: {_Xtr_name}, {_Ytr_name} (method={_method})")
```

```python

# === Correlation Heatmap — Selected Features (post-ANOVA) ===
import numpy as np, pandas as pd, matplotlib.pyplot as plt
try:
    import seaborn as sns
except Exception:
    sns = None

_selected = None
for nm in ["anova_selected_features", "selected_features", "selected_cols", "kbest_features", "fs_selected_features", "final_selected_features"]:
    if nm in globals():
        _obj = globals()[nm]
        if isinstance(_obj, (list, tuple, pd.Index)): _selected = list(_obj); break
if _selected is None and "anova_kbest" in globals():
    sk = globals()["anova_kbest"]
    if hasattr(sk, "get_support"):
        try:
            _selected = list(np.array(_dfX.columns)[sk.get_support()])
        except Exception:
            pass

if _selected is None:
    raise NameError("Could not resolve ANOVA-selected feature list. Expose it as `anova_selected_features` or similar.")

_df_base = None
for cand in ["X", "X_all", "df_features", "X_train"]:
    if cand in globals() and isinstance(globals()[cand], pd.DataFrame):
        _df_base = globals()[cand]
        break
if _df_base is None:
    raise TypeError("Could not locate a feature DataFrame to plot selected features from.")

_sel_cols = [c for c in _selected if c in _df_base.columns]
if not _sel_cols:
    raise ValueError("Selected features not found in current feature DataFrame.")
_dfSel = _df_base[_sel_cols].copy()

_corr2 = _dfSel.corr(method="spearman")
_mask2 = np.triu(np.ones_like(_corr2, dtype=bool))
plt.figure(figsize=(10,8))
if sns is not None:
    sns.heatmap(_corr2, mask=_mask2, cmap="vlag", center=0, square=False, cbar_kws={"shrink": .8})
else:
    plt.imshow(_corr2, aspect='auto', interpolation='nearest')
plt.title("Correlation Heatmap — Post-ANOVA (Selected)")
plt.tight_layout(); plt.show()
```

```python

# ==== Load fusion master ====
df = pd.read_excel(FUSION_XLSX)
print("Loaded fusion shape:", df.shape)
print("Columns:", len(df.columns))
df.head(3)
```

```python

# ==== Identify PTID and Diagnosis; clean target ====
def find_col(cols, tokens):
    for c in cols:
        lc = str(c).strip().lower()
        if any(tok in lc for tok in tokens):
            return c
    return None

ptid_col = find_col(df.columns, ["ptid","subject","rid","id"])
diag_col = find_col(df.columns, ["diagnosis","dx","label","class"])

if ptid_col is None or diag_col is None:
    raise ValueError(f"Could not find PTID/Diagnosis. Detected PTID={ptid_col}, DIAG={diag_col}")

# drop empty diagnosis
before = df.shape[0]
df = df[df[diag_col].notna()].copy()
print(f"Dropped {before - df.shape[0]} rows with empty {diag_col}")

# Normalize diagnosis labels to {CN, MCI, DEMENTIA}
def norm_diag(x):
    s = str(x).strip().upper()
    if s in {"CN", "CONTROL", "COGNITIVELY NORMAL"}:
        return "CN"
    if s in {"MCI", "EMCI", "LMCI"}:
        return "MCI"
    if s in {"AD", "DEMENTIA", "ALZHEIMER", "ALZHEIMER'S DISEASE"}:
        return "DEMENTIA"
    return s

df[diag_col] = df[diag_col].apply(norm_diag)

# Encode to 0/1/2
label_map = {"CN":0, "MCI":1, "DEMENTIA":2}
y = df[diag_col].map(label_map)
if y.isna().any():
    bad = df.loc[y.isna(), diag_col].value_counts()
    print("Unmapped diagnosis values found:\n", bad)
    raise ValueError("Please fix unmapped diagnosis labels.")

print("Class counts:\n", y.value_counts().rename(index={0:"CN",1:"MCI",2:"DEMENTIA"}))
```

```python

# ==== Build feature matrix X (drop id/target, avoid leakage dummy columns) ====
drop_like = {ptid_col, diag_col}
# also drop any one-hot remnants of diagnosis if present
drop_like |= {c for c in df.columns if str(c).strip().lower().startswith(str(diag_col).strip().lower()+"_")}

meta_cols = list(drop_like)
X = df.drop(columns=meta_cols).copy()

# Identify numeric / categorical
num_cols = [c for c in X.columns if np.issubdtype(X[c].dtype, np.number)]
cat_cols = [c for c in X.columns if c not in num_cols]

print(f"Features total: {X.shape[1]} | numeric: {len(num_cols)} | categorical: {len(cat_cols)}")
```

```python

# ==== Preprocessor + ANOVA selector ====
pre = ColumnTransformer(
    transformers=[
        ("num", SkPipeline([("imp", SimpleImputer(strategy="median")), ("sc", StandardScaler())]), num_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore", sparse_output=False), cat_cols),
    ]
)

selector = SelectKBest(score_func=f_classif, k=min(K_SELECTED,  max(1, 5 if (len(num_cols)+len(cat_cols))<5 else len(num_cols)+len(cat_cols))))
```

```python

# ==== Helpers ====
def build_pipeline(clf):
    pipe = ImbPipeline(steps=[
        ("pre", pre),
        ("adasyn", ADASYN(random_state=RANDOM_STATE)),
        ("sel", selector),
        ("clf", clf),
    ])
    return pipe

cv = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=RANDOM_STATE)

labels = ["CN","MCI","DEMENTIA"]
inv_label_map = {v:k for k,v in {"CN":0,"MCI":1,"DEMENTIA":2}.items()}
```

```python

# ==== Evaluation (CV) ====
def evaluate_model(name, base_clf):
    pipe = build_pipeline(base_clf)
    # Accuracy (macro average not needed), Macro-F1
    acc = cross_val_score(pipe, X, y, cv=cv, scoring="accuracy")
    f1m = cross_val_score(pipe, X, y, cv=cv, scoring="f1_macro")

    # ROC-AUC (OvR, macro-average). Need probabilities.
    # cross_val_predict for proba
    y_proba = cross_val_predict(pipe, X, y, cv=cv, method="predict_proba")
    try:
        auc = roc_auc_score(y, y_proba, multi_class="ovr", average="macro")
    except Exception:
        auc = np.nan

    print(f"{name}  |  Acc: {acc.mean():.3f} ± {acc.std():.3f} | Macro-F1: {f1m.mean():.3f} ± {f1m.std():.3f} | ROC-AUC: {auc:.3f}")
    # Confusion matrix via CV predictions
    y_pred = cross_val_predict(pipe, X, y, cv=cv, method="predict")
    cm = confusion_matrix(y, y_pred, labels=[0,1,2])

    # Plots
    fig, ax = plt.subplots(1,2, figsize=(12,5))
    # Confusion matrix heatmap
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False,
                xticklabels=labels, yticklabels=labels, ax=ax[0])
    ax[0].set_title(f"{name} — Confusion matrix (CV)")
    ax[0].set_xlabel("Predicted"); ax[0].set_ylabel("True")

    # ROC curves (OvR) one-vs-rest
    from sklearn.preprocessing import label_binarize
    Y_b = label_binarize(y, classes=[0,1,2])
    # Need per-fold probs; we used cross_val_predict above
    fpr, tpr = {}, {}
    for i, lab in enumerate(labels):
        from sklearn.metrics import roc_curve, auc as auc_fn
        fpr[i], tpr[i], _ = roc_curve(Y_b[:, i], y_proba[:, i])
        ax[1].plot(fpr[i], tpr[i], label=f"{lab}")
    ax[1].plot([0,1],[0,1],"k--",alpha=0.5)
    ax[1].set_title(f"{name} — ROC (OvR)")
    ax[1].set_xlabel("FPR"); ax[1].set_ylabel("TPR")
    ax[1].legend(title="Class")
    plt.tight_layout()
    plt.savefig(PLOTDIR / f"{name.lower().replace(' ','_')}_cm_roc.png", dpi=150)
    plt.show()

    # Per-class metrics table
    rep = classification_report(y, y_pred, target_names=labels, output_dict=True)
    rep_df = pd.DataFrame(rep).T
    display(rep_df)
    rep_df.to_csv(PLOTDIR / f"{name.lower().replace(' ','_')}_classification_report.csv")

    return {"model":name,
            "acc_mean":acc.mean(), "acc_std":acc.std(),
            "f1m_mean":f1m.mean(), "f1m_std":f1m.std(),
            "auc_macro_ovr":auc}
```

```python

# ==== Train & evaluate models (CV) ====
results = []

print("\n=== Random Forest ===")
rf = RandomForestClassifier(n_estimators=500, random_state=RANDOM_STATE, class_weight=None)
results.append(evaluate_model("Random Forest", rf))

if HAS_XGB:
    print("\n=== XGBoost ===")
    xgb = XGBClassifier(
        n_estimators=500,
        learning_rate=0.05,
        max_depth=4,
        subsample=0.8,
        colsample_bytree=0.8,
        objective="multi:softprob",
        eval_metric="mlogloss",
        random_state=RANDOM_STATE,
        num_class=3
    )
    results.append(evaluate_model("XGBoost", xgb))

print("\n=== SVM (RBF, prob=True) ===")
svm = SVC(kernel="rbf", probability=True, C=2.0, gamma="scale", random_state=RANDOM_STATE)
results.append(evaluate_model("SVM (RBF)", svm))

# Stacking (use simpler base models to keep runtime reasonable)
print("\n=== Stacking Ensemble ===")
base_estimators = [("rf", RandomForestClassifier(n_estimators=400, random_state=RANDOM_STATE)),
                   ("svm", SVC(kernel="rbf", probability=True, C=2.0, gamma="scale", random_state=RANDOM_STATE))]
if HAS_XGB:
    base_estimators.append(("xgb", XGBClassifier(
        n_estimators=300, learning_rate=0.05, max_depth=4,
        subsample=0.8, colsample_bytree=0.8, objective="multi:softprob",
        eval_metric="mlogloss", random_state=RANDOM_STATE, num_class=3)))

stack = StackingClassifier(
    estimators=base_estimators,
    final_estimator=LogisticRegression(max_iter=200, multi_class="ovr"),
    stack_method="predict_proba",
    passthrough=False, n_jobs=None
)

results.append(evaluate_model("Stacking", stack))

res_df = pd.DataFrame(results)
display(res_df)
res_df.to_csv(PLOTDIR / "model_cv_summary.csv", index=False)
```

```python

# ==== Fit once to get selected feature names & importances ====
def fit_get_feature_names_and_models(clf):
    pipe = build_pipeline(clf)
    pipe.fit(X, y)
    # Pull names after pre + one-hot
    feat_names = pipe.named_steps["pre"].get_feature_names_out()
    # After selector
    sel_support = pipe.named_steps["sel"].get_support()
    selected_names = list(np.array(feat_names)[sel_support])
    return pipe, selected_names

# Random Forest feature importances (if possible)
rf_pipe, rf_feats = fit_get_feature_names_and_models(RandomForestClassifier(n_estimators=500, random_state=RANDOM_STATE))
rf_clf = rf_pipe.named_steps["clf"]
if hasattr(rf_clf, "feature_importances_"):
    imp = pd.Series(rf_clf.feature_importances_, index=rf_feats).sort_values(ascending=False)
    top = imp.head(25)
    plt.figure(figsize=(8,8))
    top.iloc[::-1].plot(kind="barh")
    plt.title("Random Forest — Top 25 features (after ANOVA)")
    plt.tight_layout(); plt.savefig(PLOTDIR / "rf_top25_features.png", dpi=150); plt.show()
    imp.to_csv(PLOTDIR / "rf_feature_importances_full.csv")

# XGBoost feature importances
if HAS_XGB:
    xgb_pipe, xgb_feats = fit_get_feature_names_and_models(XGBClassifier(
        n_estimators=500, learning_rate=0.05, max_depth=4,
        subsample=0.8, colsample_bytree=0.8, objective="multi:softprob",
        eval_metric="mlogloss", random_state=RANDOM_STATE, num_class=3))
    xgb_clf = xgb_pipe.named_steps["clf"]
    if hasattr(xgb_clf, "feature_importances_"):
        imp = pd.Series(xgb_clf.feature_importances_, index=xgb_feats).sort_values(ascending=False)
        top = imp.head(25)
        plt.figure(figsize=(8,8))
        top.iloc[::-1].plot(kind="barh")
        plt.title("XGBoost — Top 25 features (after ANOVA)")
        plt.tight_layout(); plt.savefig(PLOTDIR / "xgb_top25_features.png", dpi=150); plt.show()
        imp.to_csv(PLOTDIR / "xgb_feature_importances_full.csv")
```

```python

# === Evaluation helpers (confusion, optional ROC, quick importances) ===
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix, roc_auc_score, RocCurveDisplay
from sklearn.preprocessing import label_binarize

def plot_confusion(y_true, y_pred, class_names=None, title="Confusion Matrix"):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6,5))
    plt.imshow(cm, interpolation='nearest', aspect='auto')
    plt.title(title); plt.colorbar()
    ticks = class_names if class_names is not None else [str(i) for i in range(cm.shape[0])]
    plt.xticks(range(len(ticks)), ticks, rotation=15)
    plt.yticks(range(len(ticks)), ticks)
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, cm[i, j], ha="center", va="center")
    plt.xlabel("Predicted"); plt.ylabel("True")
    plt.tight_layout(); plt.show()

def maybe_roc_auc(model, X_te, y_te):
    proba = None
    if hasattr(model, "predict_proba"):
        proba = model.predict_proba(X_te)
    elif hasattr(model, "decision_function"):
        scores = model.decision_function(X_te)
        proba = scores if scores.ndim > 1 else np.vstack([1/(1+np.exp(scores)), 1 - 1/(1+np.exp(scores))]).T
    if proba is None: 
        print("ROC: model lacks probability/decision scores."); return None
    classes = np.unique(y_te)
    y_bin = label_binarize(y_te, classes=classes)
    try:
        auc_macro = roc_auc_score(y_bin, proba, average="macro", multi_class="ovr")
        print(f"ROC AUC (macro OvR): {auc_macro:.3f}")
        if len(classes) <= 5 and hasattr(RocCurveDisplay, "from_predictions"):
            for i, cls in enumerate(classes):
                RocCurveDisplay.from_predictions((y_te==cls).astype(int), proba[:, i])
            plt.title("ROC Curves (OvR)"); plt.tight_layout(); plt.show()
        return auc_macro
    except Exception as e:
        print(f"ROC AUC failed: {e}")
        return None

def maybe_importance(model, feature_names=None, top=15, title="Top features"):
    importances = None
    if hasattr(model, "feature_importances_"):
        importances = np.asarray(model.feature_importances_)
    elif hasattr(model, "coef_"):
        coef = np.asarray(model.coef_)
        importances = np.abs(coef) if coef.ndim == 1 else np.mean(np.abs(coef), axis=0)
    if importances is None:
        print("No built-in importance; skipping."); return
    if feature_names is None or len(feature_names) != len(importances):
        feature_names = [f"f{i}" for i in range(len(importances))]
    idx = np.argsort(importances)[::-1][:top]
    plt.figure(figsize=(8, max(4, 0.35*len(idx))))
    plt.barh([feature_names[i] for i in idx[::-1]], importances[idx][::-1])
    plt.title(title); plt.tight_layout(); plt.show()
```

```python

# === Final sweep: evaluate common models if present ===
import numpy as np, pandas as pd

X_te = None; y_te = None
for nm in ["X_test_scaled", "X_test"]:
    if nm in globals(): X_te = globals()[nm]; break
if "y_test" in globals(): y_te = globals()["y_test"]
if X_te is None or y_te is None:
    raise NameError("Could not find X_test/_scaled or y_test.")

candidates = []
for nm in list(globals().keys()):
    obj = globals()[nm]
    if hasattr(obj, "predict") and hasattr(obj, "fit"):
        t = str(type(obj))
        if "sklearn" in t or "xgboost" in t:
            candidates.append(nm)

candidates = sorted(set(candidates))
summary = []

feat_names = None
for nm in ["anova_selected_features", "selected_after_anova", "selected_features"]:
    if nm in globals():
        try: feat_names = list(globals()[nm]); break
        except: pass

labels = None
for nm in ["class_names", "label_names", "diagnosis_names", "classes"]:
    if nm in globals(): labels = globals()[nm]; break

print("Detected candidate models:", candidates)
for nm in candidates:
    model = globals()[nm]
    try:
        y_pred = model.predict(X_te)
    except Exception:
        continue
    from sklearn.metrics import accuracy_score, f1_score, classification_report
    acc = accuracy_score(y_te, y_pred)
    f1m = f1_score(y_te, y_pred, average="macro")
    f1w = f1_score(y_te, y_pred, average="weighted")
    print(f"\n[{nm}] Acc={acc:.3f}  F1(macro)={f1m:.3f}  F1(weighted)={f1w:.3f}")
    print(classification_report(y_te, y_pred, digits=3))
    plot_confusion(y_te, y_pred, labels, title=f"Confusion — {nm}")
    aucm = maybe_roc_auc(model, X_te, y_te)
    maybe_importance(model, feat_names, top=15, title=f"Top features — {nm}")
    summary.append({"ModelVar": nm, "Accuracy": acc, "F1_macro": f1m, "F1_weighted": f1w, "ROC_AUC_macro": (aucm if aucm is not None else np.nan)})

if summary:
    df_sum = pd.DataFrame(summary).sort_values(["F1_macro","Accuracy"], ascending=False)
    from caas_jupyter_tools import display_dataframe_to_user
    csv_path = "/mnt/data/adni_fusion_final_balanced_model_summary.csv"
    df_sum.to_csv(csv_path, index=False)
    display_dataframe_to_user("ADNI Fusion — Final Balanced Model Summary", df_sum)
    print("Saved final summary CSV ->", csv_path)
else:
    print("No fitted models were detected to summarize.")
```
